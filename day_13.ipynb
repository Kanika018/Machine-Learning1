```
cap=cv2.VideoCapture(0)
```

```
while cap.isOpened():
  status,frame=cap.read()
  #coverting image frame into gray scale
  graying=cv2.cvtcolor(frame,cv2.COLOR_BGR2GRAY)
  print(frame.shape)
  #now we can draw all others pattrens
  #line
  cv2.line(frame,(0,0),(200,300),(0,255,0),3)
  #rectangle
  cv2.rectangle(frame,(50,50),(150,200),(255,0,0),2)
  #circle
  cv2.circle(frame,(200,300),100,(13,44,123),2)
  # text writing
  font=cv2.FONT_HERSHEY_SIMPLEX # this font type
  cv2.puttext(frame,'ADHOC',(10,50),font,2,(9,2,55),2,cv2.LINE_AA)
  
```
